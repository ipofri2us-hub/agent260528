### 1. 시스템 핵심 정의 (Definition)
본 시스템은 외부 상용 클라우드 API(OpenAI, Pinecone 등)에 전적으로 의존하는 기존 RAG 구조에서 탈피하여, 대규모 트래픽 제한 및 기업 기밀 유출 리스크를 완벽히 차단한 **100% 온프레미스(On-Premise) 완결형 교차 분석 RAG 엔진**입니다.

### 2. 주요 활용처 (Use Cases)
* **기술/트렌드 스카우팅 자동화:** 로봇 공학, 임베디드 AI 등 기술 변화 주기가 극단적으로 짧은 도메인의 글로벌 동향을 매일 아침 자동 수집 및 요약 브리핑.
* **다각도 교차 검증 리포트 발간:** 하나의 기술 키워드에 대해 국내 시장성 관점(로봇신문), 현업 개발자 벤치마크 관점(Reddit), 최신 학술 연구 관점(arXiv 논문 초록)을 단 한 장의 마크다운 리포트로 통합 출력.
* **보안 민감형 사내 지식 자산화:** 외부망이 차단된 사내 폐쇄망 환경 및 기업 미공개 기밀 정보가 포함된 연구 가이드 시스템 구축에 즉시 이식 가능.

### 3. 기존 RAG 구조와의 차이점 (Key Differences)

| 비교 항목 | 기존 일반적인 RAG 시스템 | 본 하이브리드 RAG 시스템 (my_rag_app3) |
| :--- | :--- | :--- |
| **출처 제어 (Source Balance)** | 코사인 유사도 점수 기준으로만 상위 N개를 조각내어 추출하므로, 특정 웹사이트나 도메인의 문서가 검색 결과를 전부 독점(Source Bias)하는 고질적인 문제가 발생함. | 파이썬 메모리 레벨에서 검색 결과를 재정렬하여 **3대 출처별로 각각 최대 3건씩만 할당하도록 강제하는 '균등 분할 검색(Source-Balanced Retrieval)'** 알고리즘을 구현하여 삼각 균형 시각 보장. |
| **인프라 의존도** | OpenAI, Pinecone 등 SaaS 인프라 의존으로 네트워크 지연 및 비용 리스크 존재. | **Ollama + LanceDB 기반 100% 로컬 구동**으로 운영 비용 제로화 및 기밀 유출 완벽 차단. |
| **수집 안정성 (Fallback)** | 대상 웹사이트의 스크래핑 차단(Anti-bot) 시 파이프라인 전체 마비. | Google RSS 피드 우회 및 **자가 생성형 보조 데이터(Mock Data) 맷집 설계**로 중단 없는 인덱싱. |
| **구축 오버헤드** | 무거운 독립형 인메모리 Vector DB 서버 데몬 기동 필요. | 파이썬 프로세스 내에 내장되는 파일 스트림 기반의 **초경량 LanceDB** 바인딩으로 환경 이식성 극대화. |
