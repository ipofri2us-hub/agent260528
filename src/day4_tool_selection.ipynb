# [세미나] 데이터 정의 계약 기반 로봇 정보 수집 파이프라인 아키텍처 분석

本 시스템은 로봇 공학 관련 4대 핵심 도메인(HW, ROS, AMR, RFM)의 정보 검색 파이프라인을 고정된 하드코딩 방식이 아닌, **계약 명세서(JSON)**를 중심으로 **`선택(Select) ➔ 정규화(Normalize) ➔ 검증(Validate) ➔ 실행(Execute)`**의 4단계 라이프사이클로 제어하는 모던 소프트웨어 디자인 패턴을 따르고 있습니다.


|제목|내용|설명|
|---|---|---|
|**`my_tool_contracts.json`**|소스 코드와 정책의 완전 분리 (SSOT)|허용 툴 규칙, 도메인 키워드 사전을 정적 JSON 구조에 격리하여, 파이썬 소스 코드 수정 없이 유연한 규칙 선언 및 수정이 가능합니다.|
|**`my_tool_selection.py`**|외부 장애(차단) 환경 방어 및 우회|사용자 질의 주제를 확장하여 멀티 크롤러 플랜을 빌드하며, Reddit API 봇 차단 시 Google RSS 우회 및 정적 Fallback 백업 메커니즘으로 실행 연속성을 보장합니다.|
|**`my_tool_normalizer.py`**|선 보정 후 검증을 통한 태스크 성공률 극대화|검증 전 단계에서 미인증 툴을 격리하고 누락된 인자값을 `default_values`로 자동 주입하여, 오케스트레이터의 포맷 유실 오류를 자율 보정(Self-Repair)합니다.|
|**`my_tool_validation.py`**|심각도 중심의 차등 결함 제어|최종 런타임 안정성을 심사하여 치명적 구조 결함은 `FAIL`로 프로세스를 차단하고, 단순 가이드라인 위반은 `WARNING`으로 처리해 시스템 붕괴 없이 크롤링을 지속합니다.|

|제목|내용|설명|
|---|---|---|
|**자가 치유형 파이프라인**|오류 전파 차단 및 런타임 안정성|LLM이나 룰 엔진이 불안정한 포맷을 생성하더라도, 시스템이 예외를 터트려 크래시(Crash)를 내는 대신 정규화 레이어에서 가치 판단 없이 규격화된 상태로 가공해 냅니다.|
|**인덱스 순차 재정렬 (`_renumber`)**|자료구조 레벨의 인덱스 예외 방어|파이프라인 도중 규격 외 툴이 삭제되어 실행 순서가 파편화(예: Step 1 ➔ Step 3)되더라도, 시퀀스를 1번부터 연속된 정수로 재배치하여 루프 내 바운드 에러를 원천 차단합니다.|
|**결정론적 격리 검증**|부수 효과(Side Effect) 배제와 테스트 용이성|`validate_plan`은 네트워크 상태나 외부 정답셋을 일절 참조하지 않고, 오직 데이터 컨텍스트와 JSON 계약서만 대조하므로 독립적이고 신뢰성 높은 테스트 자동화 환경을 구축할 수 있습니다.|

|제목|내용|설명|
|---|---|---|
|**Data-Driven 확장성**|비즈니스 룰 변경 시 재배포 제로화|새로운 도메인 키워드나 검색 소스가 추가될 때마다 소스 코드의 `if/else` 구문을 직접 수정하는 대신, JSON 명세만 편집하면 시스템에 동적 수용됩니다.|

|**이중 백업 탄력성**|네트워크 다운타임 극복 및 데이터 가용성|크롤러 API에서 빈 값을 반환하거나 타깃 서버가 다운되었을 때, 계약서 내 임베디드된 정적 Fallback 데이터를 싱크하여 최소한의 결과 파이프라인 출력을 지속합니다.|


각각의 비유

MCP          = USB 규격
              (어떤 기기든 같은 포트로 연결)

Tool Selection = 주문서 작성
              (메뉴판 보고 무엇을 시킬지 결정)

RAG          = 도서관 사서
              (질문에 맞는 책/자료를 찾아옴)

LangGraph    = 주방장
              (주문서 받아 순서 조율, 실패하면 다시 시도)

|제목|내용|설명|
|---|---|---|
|**툴 셀렉션 (Tool Selection)**|개념 정의|LLM이 사용자의 쿼리를 분석하여 어떤 외부 기능(API, DB 등)을 호출할지 동적으로 판단하고 매개변수를 추출하는 메커니즘|
|**RAG (검색 증강 생성)**|개념 정의 및 비교|외부 지식 데이터베이스에서 관련 문맥을 검색해 프롬프트에 주입하는 기술로, 에이전트 아키텍처의 '지식/데이터 계층'을 담당|
|**MCP (Model Context Protocol)**|개념 정의 및 비교|AI 애플리케이션과 외부 도구/데이터 간의 통신 규격을 표준화하는 프로토콜로, 에이전트 아키텍처의 '인터페이스/연결 계층'을 담당|
|**LangGraph (랭그래프)**|개념 정의 및 비교|상태(State), 노드, 엣지 구조를 기반으로 에이전트의 복잡한 작업 흐름과 피드백 루프를 설계하는 '시스템 워크플로우/제어 계층'|
|**스마트 팩토리 및 산업 자동화**|핵심 실무 응용처|LangGraph로 공정 제어 시퀀스를 설계하고, 하드웨어 센서 데이터 수집 및 비전 연산 기능을 MCP 서버로 래핑하여 에이전트에 제공|
|**시스템 트레이딩 및 퀀트 자산 관리**|핵심 실무 응용처|LangGraph로 매매 상태를 추적하고, 지표 계산 및 주문 집행은 MCP 도구로 처리하며, 거시경제 및 뉴스 분석에는 RAG를 활용|
|**엔터프라이즈 DevOps 및 QA 엔진**|핵심 실무 응용처|방대한 소스 코드와 기술 문서를 RAG에 인덱싱하여 참조하고, IDE와 컴파일러를 MCP로 연결하여 LangGraph 제어 하에 코드를 자동 수정|
